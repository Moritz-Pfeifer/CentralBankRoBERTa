```python
ecb = pd.read_pickle('/content/drive/MyDrive/CentralBankBERTa/Sentiment_classification/data_input/ECB_Sentiment_Labelled_No_Cutoff.csv')
```

```python
ecb
```

```python
ecb_positive = ecb[ecb['Predicted Sentiment'] == 1].reset_index(drop=True)
ecb_negative = ecb[ecb['Predicted Sentiment'] == 0].reset_index(drop=True)
```

```python
ecb_negative
```

```python
ecb_positive_088 = ecb_positive
index = 0 
N_length = int(len(ecb_positive_088))
for i in range(0, N_length): 
  if max(ecb_positive_088["Sentiment (Probabilities)"][index]) <= 0.88:
         ecb_positive_088['Predicted Sentiment'][index] = "n"
  index = index + 1
ecb_positive_088['Predicted Sentiment'].value_counts()
```

```python
ecb_negative_086 = ecb_negative
index = 0 
N_length = int(len(ecb_negative_086))
for i in range(0, N_length): 
  if max(ecb_negative_086["Sentiment (Probabilities)"][index]) <= 0.86:
         ecb_negative_086['Predicted Sentiment'][index] = "n"
  index = index + 1
ecb_negative_086['Predicted Sentiment'].value_counts()
```

```python
ecb_negative_086
```

```python
ecb_positive_088
```

```python
pos_ecb = ecb_positive_088[ecb_positive_088['Predicted Sentiment'] != 'n'].reset_index(drop=True)
neg_ecb = ecb_negative_086[ecb_negative_086['Predicted Sentiment'] != 'n'].reset_index(drop=True)
```

```python
neg_ecb
```

```python
pos_ecb = pos_ecb.sample(954)
neg_ecb = neg_ecb.sample(1609)
```

```python
ecb_sent_labelled = pd.concat([pos_ecb, neg_ecb])
ecb_sent_labelled
```

```python
ecb_sent_labelled = ecb_sent_labelled.drop('Sentiment (Probabilities)', axis=1)
ecb_sent_labelled = ecb_sent_labelled.rename(columns={'Predicted Sentiment': 'sentiment'})
```

```python
ecb_sent_labelled.to_csv('/content/drive/MyDrive/CentralBankBERTa/Sentiment_classification/data_input/ECB_prelabelled_sent.csv', index=0)
```
