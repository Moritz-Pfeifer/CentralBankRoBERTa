```python
print(int(len(ecb)))
ecb['Predicted Audience'].value_counts()
```

```python
households = ecb[ecb['Predicted Audience'] == 0].reset_index(drop=True)
firms = ecb[ecb['Predicted Audience'] == 1].reset_index(drop=True)
banks = ecb[ecb['Predicted Audience'] == 2].reset_index(drop=True)
government = ecb[ecb['Predicted Audience'] == 3].reset_index(drop=True)
cb = ecb[ecb['Predicted Audience'] == 4].reset_index(drop=True)
```

```python
index = 0 
N_length = int(len(households))
for i in range(0, N_length): 
  if max(households["Audience (Probabilities)"][index]) <= 0.83:
         households['Predicted Audience'][index] = "n"
  index = index + 1
households['Predicted Audience'].value_counts()
```

```python
index = 0 
N_length = int(len(firms))
for i in range(0, N_length): 
  if max(firms["Audience (Probabilities)"][index]) <= 0.65:
         firms['Predicted Audience'][index] = "n"
  index = index + 1
firms['Predicted Audience'].value_counts()
```

```python
index = 0 
N_length = int(len(banks))
for i in range(0, N_length): 
  if max(banks["Audience (Probabilities)"][index]) <= 0.92:
         banks['Predicted Audience'][index] = "n"
  index = index + 1
banks['Predicted Audience'].value_counts()
```

```python
index = 0 
N_length = int(len(government))
for i in range(0, N_length): 
  if max(government["Audience (Probabilities)"][index]) <= 0.7:
         government['Predicted Audience'][index] = "n"
  index = index + 1
government['Predicted Audience'].value_counts()
```

```python
index = 0 
N_length = int(len(cb))
for i in range(0, N_length): 
  if max(cb["Audience (Probabilities)"][index]) <= 0.77:
         cb['Predicted Audience'][index] = "n"
  index = index + 1
cb['Predicted Audience'].value_counts()
```

```python
h = households[households['Predicted Audience'] != 'n'].reset_index(drop=True)
f = firms[firms['Predicted Audience'] != 'n'].reset_index(drop=True)
b = banks[banks['Predicted Audience'] != 'n'].reset_index(drop=True)
g = government[government['Predicted Audience'] != 'n'].reset_index(drop=True)
cb = cb[cb['Predicted Audience'] != 'n'].reset_index(drop=True)
```

```python
# take samples here
g_ecb = g.sample(624)
cb_ecb = cb.sample(704)
```

```python
ecb_labelled = pd.concat([h,f,b,g_ecb,cb_ecb]).reset_index(drop=True)
```

```python
#ecb_labelled = ecb_labelled.drop(['Audience (Probabilities)'], axis=1)
ecb_labelled = ecb_labelled.rename(columns={'text': 'Parsed_Text', 'Predicted Audience': 'label'})
ecb_labelled
```
